```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(123)

n_sample=1000
class_0_ratio=0.9
n_class_0=int(n_sample * class_0_ratio)
n_class_1=n_sample -n_class_0
```

```python
n_class_0,n_class_1
```

```python
class_0=pd.DataFrame({"feature1":np.random.normal(loc=0,scale=1,size=n_class_0),
                      "feature2":np.random.normal(loc=0,scale=1,size=n_class_0),
                      "target":[0]*n_class_0}
                     )
```

```python
class_1=pd.DataFrame({"feature1":np.random.normal(loc=0,scale=1,size=n_class_1),
                      "feature2":np.random.normal(loc=0,scale=1,size=n_class_1),
                      "target":[1]*n_class_1}
                     )
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
df['target'].value_counts()
```

Upsampling technique

make lowest count data to match with higest
ex. make minority(100 records) to match with mah=jority(900)
100=>900

```python
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
type(df_minority)
```

```python
#perform upsampling
```

```python
from sklearn.utils import resample
```

```python
df_minority_upsample=resample(df_minority
                              ,replace=True   #sample with replacement
                              ,n_samples=len(df_majority)  #to match with majority class
                              ,random_state=43  #use any number here
                              )
df_minority_upsample.shape
```

```python
df_upsampled=pd.concat([df_minority_upsample,df_majority])
df_upsampled['target'].value_counts()
```

Downsampling

make higest count data to match with lowest
ex. make majority(900 records) to match with minority(100)
900=>100


```python
class_0=pd.DataFrame({"feature1":np.random.normal(loc=0,scale=1,size=n_class_0),
                      "feature2":np.random.normal(loc=0,scale=1,size=n_class_0),
                      "target":[0]*n_class_0}
                     )

class_1=pd.DataFrame({"feature1":np.random.normal(loc=0,scale=1,size=n_class_1),
                      "feature2":np.random.normal(loc=0,scale=1,size=n_class_1),
                      "target":[1]*n_class_1}
                     )

df=pd.concat([class_0,class_1]).reset_index(drop=True)

df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
df_majority_downsample=resample(df_majority
                              ,replace=False   #sample without replacement
                              ,n_samples=len(df_minority)  #to match with majority class
                              ,random_state=43  #use any number here
                              )
```

```python
df_downsampled=pd.concat([df_majority_downsample,df_minority])
df_downsampled['target'].value_counts()
```
